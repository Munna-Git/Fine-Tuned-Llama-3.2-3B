```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# CELL 1: Install Dependencies (SAME AS BEFORE)
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

import sys
import subprocess

print("ğŸ§¹ Cleaning existing installations...")
packages_to_remove = ['transformers', 'accelerate', 'peft', 'bitsandbytes', 'trl', 'datasets']
for pkg in packages_to_remove:
    subprocess.run([sys.executable, '-m', 'pip', 'uninstall', '-y', pkg], 
                   stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

print("ğŸ“¦ Installing compatible libraries...")
subprocess.run([sys.executable, '-m', 'pip', 'install', '-q', 'transformers==4.46.3'])
subprocess.run([sys.executable, '-m', 'pip', 'install', '-q', 'accelerate==1.11.0'])
subprocess.run([sys.executable, '-m', 'pip', 'install', '-q', 'peft==0.13.2'])
subprocess.run([sys.executable, '-m', 'pip', 'install', '-q', 'bitsandbytes==0.45.5'])
subprocess.run([sys.executable, '-m', 'pip', 'install', '-q', 'trl==0.11.4'])
subprocess.run([sys.executable, '-m', 'pip', 'install', '-q', 'datasets==3.0.1'])

print("âœ… Core libraries installed!")

import os
os.environ['TOKENIZERS_PARALLELISM'] = 'false'
os.environ['OMP_NUM_THREADS'] = '2'
os.environ['MKL_NUM_THREADS'] = '2'
os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'max_split_size_mb:512'

import torch
import transformers

print(f"\\nğŸ”§ System Status:")
print(f"   PyTorch: {torch.__version__}")
print(f"   Transformers: {transformers.__version__}")
print(f"   CUDA Available: {torch.cuda.is_available()}")

if torch.cuda.is_available():
    num_gpus = torch.cuda.device_count()
    print(f"   GPUs Detected: {num_gpus}")
    for i in range(num_gpus):
        props = torch.cuda.get_device_properties(i)
        print(f"   GPU {i}: {props.name}")
        print(f"      Memory: {props.total_memory / 1e9:.2f} GB")
    
    torch.cuda.empty_cache()
    print(f"\\nâœ… Multi-GPU Training Ready ({num_gpus} GPUs)")
```

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# CELL 2: ğŸ”§ FIX #1 - IMPROVED DATA LOADING WITH FORMAT VALIDATION
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

import json
from datasets import Dataset
import gc

TRAIN_PATH = "/kaggle/input/acord-data/train.json"
VAL_PATH = "/kaggle/input/acord-data/val.json"

print("ğŸ“‚ Loading and validating training data...")

with open(TRAIN_PATH, 'r') as f:
    train_data = json.load(f)
with open(VAL_PATH, 'r') as f:
    val_data = json.load(f)

print(f"âœ… Data loaded: {len(train_data)} train, {len(val_data)} val samples")

# ğŸ” DATA VALIDATION - Check if outputs are complete clauses or summaries
print("\\nğŸ” Validating data quality...")
sample_outputs = [ex['output'][:100] for ex in train_data[:5]]
for i, output in enumerate(sample_outputs, 1):
    print(f"   Sample {i} output preview: {output}...")

avg_output_len = sum(len(ex['output']) for ex in train_data) / len(train_data)
print(f"   Average output length: {avg_output_len:.0f} characters")
if avg_output_len < 50:
    print("   âš ï¸ WARNING: Outputs seem short - may be summaries not full clauses!")
else:
    print("   âœ… Output lengths look reasonable")
```

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# CELL 3: ğŸ”§ FIX #2 - CORRECTED INSTRUCTION FORMAT FOR EXTRACTION
# This teaches the model to COPY text, not explain it
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

def format_instruction_fixed(example):
    """"
    FIXED FORMAT: Uses direct instruction that encourages extraction, not explanation.
    
    Changes from original:
    1. Simpler, more direct instruction
    2. Emphasizes 'copy the exact text'
    3. Removed conversational system message
    4. Shorter prompt to reduce verbosity
    """
    
    # Extract clause type from instruction
    clause_type = example.get('clause_type', 'requested clause')
    
    # CRITICAL FIX: Direct extraction instruction, not conversational
    text = (
        f"<|begin_of_text|><|start_header_id|>user<|end_header_id|>\\n\\n"
        f"Copy the exact '{clause_type}' clause from this contract:\\n\\n"
        f"{example['input'][:1500]}<|eot_id|>"  # Limit input to avoid truncation
        f"<|start_header_id|>assistant<|end_header_id|>\\n\\n"
        f"{example['output']}<|eot_id|>"
    )
    
    return {"text": text}

print("ğŸ”„ Formatting datasets with FIXED prompt template...")

train_formatted = [format_instruction_fixed(ex) for ex in train_data]
val_formatted = [format_instruction_fixed(ex) for ex in val_data]

train_dataset = Dataset.from_list(train_formatted)
val_dataset = Dataset.from_list(val_formatted)

del train_data, val_data, train_formatted, val_formatted
gc.collect()

print(f"âœ… Datasets ready!")
print(f"   Train: {len(train_dataset)} samples")
print(f"   Validation: {len(val_dataset)} samples")
print(f"\\nExample formatted prompt (first 400 chars):\\n{train_dataset[0]['text'][:400]}...\\n")

```

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# CELL 4: Load Model (SAME AS BEFORE)
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
import torch
from huggingface_hub import login
import os

# Hugging Face authentication
os.environ["HF_TOKEN"] = HF_TOKEN # Replace with your token
login(token=os.environ["HF_TOKEN"])
print("âœ… Hugging Face authentication successful!")

MODEL_NAME = "meta-llama/Llama-3.2-3B-Instruct"
MAX_SEQ_LENGTH = 2048

print(f"ğŸ“¥ Loading {MODEL_NAME}...")

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_use_double_quant=True,
)

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right"

model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    quantization_config=bnb_config,
    device_map="auto",
    torch_dtype=torch.float16,
    trust_remote_code=True,
    low_cpu_mem_usage=True,
)

print(f"âœ… Model loaded!")
print(f"   Memory: {model.get_memory_footprint() / 1e9:.2f} GB")

model.config.use_cache = False
model.config.pretraining_tp = 1
torch.cuda.empty_cache()
```

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# CELL 5: Apply LoRA (SAME AS BEFORE)
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training

print("ğŸ”§ Preparing model for LoRA training...")

model.gradient_checkpointing_enable()
model = prepare_model_for_kbit_training(model)

lora_config = LoraConfig(
    r=32,
    lora_alpha=64,
    target_modules=[
        "q_proj", "k_proj", "v_proj", "o_proj",
        "gate_proj", "up_proj", "down_proj",
    ],
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM",
)

model = get_peft_model(model, lora_config)

trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
total_params = sum(p.numel() for p in model.parameters())

print(f"âœ… LoRA applied!")
print(f"   Trainable parameters: {trainable_params:,}")
print(f"   Trainable percentage: {100 * trainable_params / total_params:.4f}%")

```

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# CELL 6: ğŸ”§ FIX #3 - IMPROVED TRAINING CONFIGURATION
# Lower learning rate, more logging, better optimization
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

from transformers import TrainingArguments

OUTPUT_DIR = "./legal-llm-output"
num_gpus = torch.cuda.device_count()

print(f"ğŸ¯ Configuring training for {num_gpus} GPU(s)...")

# ğŸ”§ CRITICAL FIX: Lower learning rate for better convergence
if num_gpus == 2:
    per_device_batch = 2
    grad_accum = 4
else:
    per_device_batch = 2
    grad_accum = 8

training_args = TrainingArguments(
    output_dir=OUTPUT_DIR,
    overwrite_output_dir=True,
    
    # Training
    num_train_epochs=3,
    per_device_train_batch_size=per_device_batch,
    gradient_accumulation_steps=grad_accum,
    per_device_eval_batch_size=1,
    
    # ğŸ”§ FIX: Lower learning rate for better convergence
    learning_rate=1e-4,  # REDUCED from 2e-4
    weight_decay=0.01,
    warmup_steps=100,    # INCREASED warmup
    max_grad_norm=0.3,   # REDUCED for stability
    optim="paged_adamw_8bit",  # More memory efficient
    
    # Precision
    fp16=True,
    bf16=False,
    
    # CPU optimization
    dataloader_num_workers=0,
    dataloader_pin_memory=False,
    
    # Disabled features
    torch_compile=False,
    gradient_checkpointing=True,
    gradient_checkpointing_kwargs={"use_reentrant": False},
    
    # Multi-GPU
    ddp_find_unused_parameters=False,
    
    # ğŸ”§ FIX: More frequent logging to track progress
    logging_steps=10,    # INCREASED logging frequency
    logging_strategy="steps",
    logging_first_step=True,
    
    # Evaluation
    eval_strategy="steps",
    eval_steps=100,
    eval_on_start=False,
    
    # Saving
    save_strategy="steps",
    save_steps=200,
    save_total_limit=2,
    load_best_model_at_end=False,
    
    # Other
    report_to="none",
    seed=42,
    data_seed=42,
    remove_unused_columns=False,
    label_names=["labels"],
)

print(f"âœ… Training configuration complete!")
print(f"   Learning rate: {training_args.learning_rate}")
print(f"   Effective batch size: {per_device_batch * num_gpus * grad_accum}")
print(f"   Warmup steps: {training_args.warmup_steps}")

```

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# CELL 7: Train Model (SAME AS BEFORE)
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

from trl import SFTTrainer
import glob

print("ğŸš€ Initializing trainer...")

checkpoint_dirs = sorted(glob.glob(f"{OUTPUT_DIR}/checkpoint-*"))
resume_checkpoint = checkpoint_dirs[-1] if checkpoint_dirs else None

if resume_checkpoint:
    print(f"ğŸ“‚ Resuming from: {resume_checkpoint}")
else:
    print(f"ğŸ“‚ Starting from scratch")

torch.cuda.empty_cache()

trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    dataset_text_field="text",
    max_seq_length=MAX_SEQ_LENGTH,
    args=training_args,
    packing=True,
)

print(f"âœ… Trainer initialized!")
print(f"\\n{'='*70}")
print(f"ğŸ¯ STARTING TRAINING")
print(f"{'='*70}\\n")

try:
    train_result = trainer.train(resume_from_checkpoint=resume_checkpoint)
    
    print(f"\\n{'='*70}")
    print(f"âœ… TRAINING COMPLETE!")
    print(f"{'='*70}")
    print(f"   Train Loss: {train_result.training_loss:.4f}")
    print(f"   Time: {train_result.metrics['train_runtime']/60:.1f} minutes")
    
except Exception as e:
    print(f"âŒ Training error: {e}")
    raise

torch.cuda.empty_cache()

```

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# CELL 8: Save Model (SAME AS BEFORE)
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

print("ğŸ’¾ Saving fine-tuned model...")

lora_path = f"{OUTPUT_DIR}/lora_adapters"
os.makedirs(lora_path, exist_ok=True)

model.save_pretrained(lora_path)
tokenizer.save_pretrained(lora_path)

print(f"âœ… LoRA adapters saved to: {lora_path}")
```

```python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# CELL 9: ğŸ”§ FIX #4 - CORRECT EVALUATION USING PEFT MODEL
# This fixes the 0 trainable params issue
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

import random
from difflib import SequenceMatcher
from tqdm import tqdm

print("ğŸ”„ PREPARING FOR EVALUATION")
print("="*70)

# ğŸ”§ CRITICAL FIX: Keep model with LoRA adapters for evaluation
# DO NOT call merge_and_unload() before evaluation!

# Set to evaluation mode
model.eval()

# Verify model has trainable parameters
trainable = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"Trainable parameters: {trainable:,}")
if trainable > 0:
    print("âœ… Model has LoRA adapters attached")
else:
    print("âŒ WARNING: No trainable params - model may not be fine-tuned!")

def similarity_score(s1, s2):
    return SequenceMatcher(None, s1.lower(), s2.lower()).ratio()

# Load test data
TEST_PATH = "/kaggle/input/acord-data/test.json"
with open(TEST_PATH, 'r') as f:
    test_data = json.load(f)

print(f"\\nEvaluating on 100 test samples...")
print("="*70)

random.seed(42)
eval_samples = random.sample(test_data, min(100, len(test_data)))

results = []
excellent = 0
good = 0
partial = 0
similarities = []

for i, sample in enumerate(tqdm(eval_samples, desc="Evaluating")):
    try:
        # ğŸ”§ FIX: Use same format as training for consistency
        prompt = (
            f"<|begin_of_text|><|start_header_id|>user<|end_header_id|>\\n\\n"
            f"Copy the exact '{sample.get('clause_type', 'requested clause')}' clause from this contract:\\n\\n"
            f"{sample['input'][:1500]}<|eot_id|>"
            f"<|start_header_id|>assistant<|end_header_id|>\\n\\n"
        )
        
        inputs = tokenizer(prompt, return_tensors="pt", truncation=True, max_length=2000)
        inputs = {k: v.to(model.device) for k, v in inputs.items()}
        
        with torch.no_grad():
            outputs = model.generate(
                **inputs,
                max_new_tokens=150,
                temperature=0.1,
                do_sample=False,  # Greedy for consistency
                pad_token_id=tokenizer.eos_token_id,
            )
        
        generated = tokenizer.decode(outputs[0], skip_special_tokens=True)
        
        # Extract answer after assistant
        if "assistant" in generated:
            predicted = generated.split("assistant")[-1].strip()
        else:
            predicted = generated.strip()
        
        # Clean prediction
        predicted = predicted.split("<|eot_id|>")[0].strip()
        actual = sample['output']
        
        # Calculate similarity
        sim = similarity_score(predicted, actual)
        similarities.append(sim)
        
        if sim > 0.9:
            excellent += 1
        if sim > 0.7:
            good += 1
        if sim > 0.5:
            partial += 1
        
        results.append({
            'clause_type': sample.get('clause_type', 'unknown'),
            'predicted': predicted[:200],
            'actual': actual[:200],
            'similarity': sim,
        })
        
        # Show first 3 for debugging
        if i < 3:
            print(f"\\nSample {i+1}: {sample.get('clause_type', 'unknown')}")
            print(f"   Predicted: {predicted[:100]}...")
            print(f"   Actual: {actual[:100]}...")
            print(f"   Similarity: {sim:.1%}")
    
    except Exception as e:
        print(f"\\nError on sample {i}: {e}")
        continue

# Print results
n = len(results)
print("\\n" + "="*70)
print("ğŸ“Š EVALUATION RESULTS")
print("="*70)
print(f"Samples: {n}")
print(f"Excellent (>90%): {excellent/n:.1%} ({excellent}/{n})")
print(f"Good (>70%): {good/n:.1%} ({good}/{n})")
print(f"Partial (>50%): {partial/n:.1%} ({partial}/{n})")
print(f"Average Similarity: {sum(similarities)/n:.1%}")
print("="*70)

# Save results
import json
with open(f'{OUTPUT_DIR}/evaluation_results.json', 'w') as f:
    json.dump(results, f, indent=2)

print(f"\\nâœ… Results saved to: {OUTPUT_DIR}/evaluation_results.json")

# ğŸ‰ SUCCESS METRICS
if sum(similarities)/n > 0.5:
    print("\\nğŸ‰ SUCCESS! Model shows significant improvement over baseline!")
elif sum(similarities)/n > 0.3:
    print("\\nâš ï¸ MODERATE: Model learned some patterns but needs improvement")
else:
    print("\\nâŒ NEEDS WORK: Check data quality and training hyperparameters")


# Save to file
with open('/tmp/fixed_week2_training.txt', 'w') as f:
    f.write(fixed_training_code)

print("âœ… Fixed training code generated!")
print(f"   Total lines: {len(fixed_training_code.splitlines())}")
print(f"\\nKey fixes implemented:")
print("   1. âœ… Simplified instruction format for extraction")
print("   2. âœ… Data validation checks")
print("   3. âœ… Lower learning rate (1e-4 vs 2e-4)")
print("   4. âœ… Evaluation uses PEFT model (not merged)")
print("   5. âœ… Consistent prompt format in train/eval")
print("   6. âœ… More logging and diagnostics")
```
